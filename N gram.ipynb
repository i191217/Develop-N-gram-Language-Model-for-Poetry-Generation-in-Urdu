```python
import spacy
import pandas as pd
import random

sp = spacy.load('en_core_web_sm')
```

```python
li = []
str1 = " "

all_stopwords = sp.Defaults.stop_words
all_stopwords |= {"'","’",'“','Wafa','Thay','Tha','Rastay','Rahay','Musafir','Muqaddar','Muhabbat','Main','Kiray','K','Ghar','Chaltay','Chalna','Berukhi','Badalty','Badaltay','Adawat',':','.',"'","%","آ","۔","ء","-","و",'’’',"‘",'"',")","(","*","!","،","٪","'","؟"}
```

```python
def to_csv(filename, csv_name):
  df = pd.read_table('/content/drive/My Drive/Colab Notebooks/files/' + filename, names = ['Words', 'Probabilities'], index_col=0)
  df.to_csv('/content/drive/My Drive/Colab Notebooks/files/' + csv_name)
  df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/files/' + csv_name)
  df = df.groupby('Words').size().div(len(df))
  df.to_csv('/content/drive/My Drive/Colab Notebooks/files/' + csv_name)
  df2 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/files/' + csv_name)
  df2.columns = ['Words', 'Probabilities']
  df2 = df2.drop_duplicates(subset = ["Words"])
  df2.to_csv('/content/drive/My Drive/Colab Notebooks/files/' + csv_name, index=False)
```

**Uni-gram Model**

```python
f = open("/content/drive/My Drive/Colab Notebooks/files/faiz + iqbal + ghalib.txt",encoding="utf8")
# f = open("/content/drive/My Drive/Colab Notebooks/files/15.txt",encoding="utf8")
a = f.readlines()

for line in a:
  doc = sp(line)
  for token in doc:
      li.append(token.text)

  tokens_without_sw = [word for word in li if not word in all_stopwords]
  li.clear()    

  data = str1.join(tokens_without_sw)
  data = sp(data)

  for wc in data:
      if wc.text != "\n":
          fg = open("/content/drive/My Drive/Colab Notebooks/files/uni-gram.txt", "a", encoding="utf8")
          # fg = open("/content/drive/My Drive/Colab Notebooks/files/15U.txt", "a", encoding="utf8")
          fg.write(wc.text + "\n")
          # print(wc.text)
```

**Bi-gram Model**

```python
f = open("/content/drive/My Drive/Colab Notebooks/files/faiz + iqbal + ghalib.txt",encoding="utf8")
# f = open("/content/drive/My Drive/Colab Notebooks/files/15.txt",encoding="utf8")
a = f.readlines()

for line in a:
  doc = sp(line)
  for token in doc:
      li.append(token.text)

  tokens_without_sw = [word for word in li if not word in all_stopwords]
  li.clear()    

  for i in range(len(tokens_without_sw)-2):
    if tokens_without_sw[i] and tokens_without_sw[i + 1] != "\n":
      # print(tokens_without_sw[i] + " " + tokens_without_sw[i + 1])
      fg = open("/content/drive/My Drive/Colab Notebooks/files/bi-gram.txt", "a", encoding="utf8")
      fg.write(tokens_without_sw[i] + " " + tokens_without_sw[i + 1] + "\n")

print("End")
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/files/sample.csv')
largest = df.nlargest(1000,'Probabilities')
print(type(largest))
print(largest)
```

```python
# print(largest['Words'])
# print(" ")

final = []
final2 = []
final3 = []
list1 = largest['Words'].values.tolist()
# print(list1)


trigram = []
sample3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/files/sample3.csv')
list3 = sample3.values.tolist()
# print(list3)
# print(list3[0])
# print(list3[0][0])

for x3 in range(3):
  for x2 in range(4):
    rand_word = random.choice(list1)
    for x in range(3):
      print(rand_word)
      for i in list3:
        ss = sp(i[0])
        for h in ss:
          if str(h) == rand_word:
            trigram.append(i)
          break
          
      print("Trigram:",trigram)
      max = trigram[0][1]
      append = trigram[0][0]
      print("max:",max)

      for i in trigram:
        if i[1] > max:
          max = i[1]
          append = i[0]

      print("max2:",max)
      print("Append:",append)

      # append[1st 2 words]
      print(len(append.split()))

      dc = sp(append)
      rand_word = str(dc[-1])
      trigram.clear()

      if len(final) < 6:
        if len(final) == 4:
          for k in dc:
            final.append(k.text)
        else:
          for k in dc[0:2]:
            final.append(k.text)
        
      final2 = final.copy()
      final3.append(final2)
      final.clear()

print("Final list:",final3)
```
